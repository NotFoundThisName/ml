```python
import numpy as np
```

```python
#定义数据
# 自变量，每周学习时长
X = [[5], [8], [10], [12], [15], [3], [7], [9], [14], [6]]
# 因变量，数学考试成绩
y = [55, 65, 70, 75, 85, 50, 60, 72, 80, 58]
```

```python
X = np.array(X)
y = np.array(y).reshape(-1, 1)

#X 增加一列
n = X.shape[0]
X = np.hstack((np.ones((n, 1)), X)) #水平堆叠 第一列是一列1，第二列是X
```

```python
#定义损失函数
def J(beta):
    return np.sum((X @ beta - y) ** 2) / n
```

```python
#定义梯度计算方法
def gradient(beta):
    return  X.T @ (X @ beta - y) / n * 2
```

```python
#1.初始化参数和超参数
bate = np.array([1,1]).reshape(2, 1)
alpha = 0.01
iter = 10000 #迭代次数

bate
```
