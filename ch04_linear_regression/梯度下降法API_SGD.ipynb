```python
from sklearn.linear_model import SGDRegressor
```

```python
#定义数据
# 自变量，每周学习时长
X = [[5], [8], [10], [12], [15], [3], [7], [9], [14], [6]]
# 因变量，数学考试成绩
y = [55, 65, 70, 75, 85, 50, 60, 72, 80, 58]
```

```python
#2.定义模型
model = SGDRegressor(
    loss='squared_error', #使用均方误差 损失函数
    learning_rate='constant', #学习率选择 constant代表是常数
    fit_intercept=True, #计算截距
    max_iter=10 ** 6 ,
    eta0=1e-4,#学习率
    tol=1e-6 #损失率变化量比这个小就会停止
)
```

```python
#训练模型
model.fit(X, y)
```

```python
#4.打印参数
model.coef_
```
