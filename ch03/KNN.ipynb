```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#定义数据
X = [[2,1],[3,1],[1,4],[2,6]] # 特征
y = [0,0,1,1] #标签
```

```python
#定义模型 KNN分类器
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
#训练
knn.fit(X,y)
```

```python
#预测
x = [[4,9]]
x_class = knn.predict(x)
x 
```

```python
#画图
import matplotlib.pyplot as plt
import numpy as np
```

```python
#转成np的ndarry
X = np.array(X)
y = np.array(y)
```

```python
#bool索引 取出前两个点
X1 = X[y == 0]
X2 = X[y == 1]
X1
```

```python
colors = ['C0','C1']
plt.scatter(X1[:,0],X1[:,1],color = colors[0])
plt.scatter(X2[:,0],X2[:,1],color = colors[1])

#绘制新的样本点
x_color = colors[0] if x_class == 0 else colors[1]
plt.scatter(x[0][0],x[0][1],c = x_color)
plt.show()
```
