```python
from sklearn.preprocessing import MinMaxScaler
#定义数据
X = [[2,1],[3,1],[1,4],[2,6]] # 特征
```

```python
#定义归一化缩放器
scaler = MinMaxScaler()
```

```python
X = scaler.fit_transform(X)
X
```

```python
from sklearn.preprocessing import StandardScaler
#定义数据
X = [[2,1],[3,1],[1,4],[2,6]] # 特征
```

```python
#定义标准化缩放器
scaler = StandardScaler()
```

```python
#标准化处理
X = scaler.fit_transform(X)
X
```
