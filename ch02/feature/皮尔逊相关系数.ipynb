```python
import pandas as pd
```

```python
#读取数据
data = pd.read_csv('../../data/advertising.csv')
print(data)
```

```python
#数据预处理,去掉第一列
data.dropna(inplace=True)
data.drop(data.columns[0],axis=1,inplace=True)
data
```

```python
#定义特征x和标签y
x = data.drop("Sales",axis=1)
y = data["Sales"]
print(x)
print(y)
```

```python
#计算皮尔逊相关系数
r = x.corrwith(y,method="pearson")
r
```

```python
#计算皮尔逊相关系数矩阵
corr_maxtrix = data.corr(method='pearson')
corr_maxtrix
```

```python
#可视化展示
import seaborn as sns
sns.heatmap(corr_maxtrix,annot=True,fmt=".2f",cmap="coolwarm")
```
