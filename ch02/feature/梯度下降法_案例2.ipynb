```python
import numpy as np
#构造目标函数
def J(x):
    return (x ** 2 - 2) ** 2

#梯度函数
def gradient(x):
    return 4 * x ** 3 - 8 * x
```

```python
#用列表保存点的变化轨迹
y_list = []
x_list = []
#1. 初始化参数（x）和超参数
x = 1
alpha = 0.1
#4. 重复迭代100次,当梯度接近0时停止
while np.abs(grad := gradient(x)) > 1e-10:
    #记录当前点的坐标
    x_list.append(x)
    y_list.append(J(x))

    # 2. 计算梯度
    #grad = gradient(x)

    # 3. 更新参数
    x = x - alpha * grad
print(len(x_list))
```

```python
#画图
import matplotlib.pyplot as plt
import numpy as np

x= np.arange(0.9,1.6,0.01)
plt.plot(x,J(x))
plt.plot(x_list,y_list)
plt.scatter(x_list,y_list,c = 'r')
plt.show()
```
