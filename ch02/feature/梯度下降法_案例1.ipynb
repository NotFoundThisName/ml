```python
#定义目标函数 （模拟损失函数）
def f(x):
        return x ** 2

#定义导函数 （计算梯度的函数）
def gradint(x):
    return 2 * x
```

```python
#用列表保存点的变化轨迹
y_list = []
x_list = []
```

```python
#1. 初始化参数（x）和超参数
x = 1
alpha = 0.1
```

```python
#4. 重复迭代100次
for i in range(100):
    #记录当前点的坐标
    x_list.append(x)
    y_list.append(f(x))
    print(x_list[i], y_list[i])
    # 2. 计算梯度
    grad = gradint(x)
    
    # 3. 更新参数
    x= x- alpha * grad
```

```python
#画图
import matplotlib.pyplot as plt
import numpy as np

x= np.arange(-1,1,0.01)
plt.plot(x,f(x))
plt.plot(x_list,y_list)
plt.scatter(x_list,y_list,c = 'r')
plt.show()
```
