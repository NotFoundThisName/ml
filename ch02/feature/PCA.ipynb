```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
X = np.random.randn(1000,3)
X
```

```python
#定义PCA对象
pca = PCA(n_components=2)
```

```python
#应用PCA，降维处理
X_pca = pca.fit_transform(X)
X_pca
```

```python
#画图
```

```python
fig = plt.figure(figsize=(12,4))

#添加子图 3D数据可视化 121指的是一行两列中的第一个
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X[:,0], X[:,1], X[:,2], c='g')
ax1.set_title('Before PCA (3D)')
ax1.set_xlabel('Fature1')
ax1.set_ylabel('Fature2')
ax1.set_zlabel('Fature3')

ax2 = fig.add_subplot(122)
ax2.scatter(X_pca[:,0], X_pca[:,1], c='blue')
ax2.set_title('Before PCA (2D)')
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
```

```python
#定义本质特征（主成分）
n = 1000
pc1 = np.random.normal(0,1,n)
pc2 = np.random.normal(0,0.2,n)
noice = np.random.normal(0,0.1,n)

#手动构造特征数据(基于主成分做叠加组合)
x1 = pc1 - pc2
x2 = pc1 + pc2
x3 = pc2 + noice

x = np.vstack((x1,x2,x3)).T
x
```

```python
#定义PCA对象
pca = PCA(n_components=2)
#应用PCA，降维处理
x_pca = pca.fit_transform(x)
x_pca
```

```python
fig = plt.figure(figsize=(12,4))

#添加子图 3D数据可视化 121指的是一行两列中的第一个
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(x[:,0], x[:,1], x[:,2], c='g')
ax1.set_title('Before PCA (3D)')
ax1.set_xlabel('Fature1')
ax1.set_ylabel('Fature2')
ax1.set_zlabel('Fature3')

ax2 = fig.add_subplot(122)
ax2.scatter(x_pca[:,0], x_pca[:,1], c='blue')
ax2.set_title('Before PCA (2D)')
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
```
