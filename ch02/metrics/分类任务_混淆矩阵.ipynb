```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score , precision_score, recall_score, f1_score
```

```python
label = ["猫","狗"] #标签
y_true = ["猫","猫","猫","狗","猫","狗","猫","狗","猫","狗"]
y_pred = ["猫","猫","猫","猫","猫","狗","猫","狗","猫","狗"]
```

```python
matrix = confusion_matrix(y_true, y_pred)
matrix
```

```python
accuracy = accuracy_score(y_true, y_pred)
accuracy
```

```python
precision_score(y_true,y_pred,pos_label="猫")
```

```python
recall_score(y_true,y_pred,pos_label="猫")
```

```python
f1 = f1_score(y_true,y_pred,pos_label="猫")
f1
```
