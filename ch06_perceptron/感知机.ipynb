```python
def AND(x1,x2):
    w1,w2,theta = 0.5,0.5,0.7
    if w1*x1 + w2*x2 >= theta:
        return 1
    else:
        return 0
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
import numpy as np
def AND1(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.7
    res = w @ x + b
    if res >= 0:
        return 1
    else:
        return 0
```

```python
print(AND1(0,0))
print(AND1(0,1))
print(AND1(1,0))
print(AND1(1,1))
```
